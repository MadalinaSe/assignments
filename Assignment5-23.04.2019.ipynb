## Read data

## Merge GDP and Pop

```r
data<- full_join(GDP, Pop, by=c("Year","Country"))
str(data)
data$GDP <- as.numeric(as.character(data$GDP))
data$Pop<- as.numeric(as.character(data$Pop))
str(data)
head(data)
```

```r
GDP_cap<- data$GDP/data$Population
data_col<- cbind(data, GDP_cap)
```

```r
final <- filter(data_col, Country == "Australia" | Country=="Netherlands" | Country=="United Kingdom"|Country=="Germany")
final
```

```r
ggplot(final)+
    geom_line(aes(x=Year, y=GDP_cap, color=Country))+
    ylab("GDP oer capita")

ggsave("../../output/gdp_cap.png")
```
